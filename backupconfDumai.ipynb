```python
import telnetlib
import time
import pandas as pd
import paramiko
import sys
```

```python
devices  = pd.read_excel("AllDevices.xlsx",sheet_name="Dumai",header=2)
ips = devices["NEW IP RTN"]
devicex= devices["NAMA DEVICE"]
```

```python
def telnet(ip,device):
    user = b"admin01\r\n"
    password = b"C1sc0!234\r\n"
    HOST = ip
    tn = telnetlib.Telnet(HOST)
    tn.read_until("Username: ".encode('utf-8'))
    tn.write(user)
    tn.read_until("Password: ".encode('utf-8'))
    tn.write(password)
    #tn.write("vt100\n")
    tn.write(b"ter len 0\r\n")
    tn.write(b"sh run\r\n")
    tn.write(b"exit\r\n")
    out = tn.read_all().decode('ascii')
    filen= open(r"Dumai\{}.txt".format(device), 'w')
    filen.write(out)
    filen.close()

```

```python
def sshx(ip,device):
    router_ip = ip
    router_username = "admin01"
    router_password = "C1sc0!234"
    user = router_username
    password = router_password
    port=22
    ssh = paramiko.SSHClient()
    ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
    ssh.connect(router_ip,port, user, password, look_for_keys=False)
    chan = ssh.invoke_shell()
    time.sleep(1)
    chan.send('term len 0\n')
    time.sleep(1)
    chan.send('sh run\n')
    time.sleep(2)
    output = chan.recv(999999)
    f1 = open(r"Dumai\{}.txt".format(device), 'w')
    f1.write(output.decode("utf-8") )
    f1.close()
    ssh.close() 
```

```python
for (ip,device) in zip(ips,devicex):
    if ip != ip:
        continue
    try:
        telnet(ip,device)
    except:
        try:
            sshx(ip,device)
        except:
            continue
```
